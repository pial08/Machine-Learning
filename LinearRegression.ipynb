# Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np

salary = np.array([20,30,40,50,60])
food = np.array([5,10,15,20,25])

plt.xlabel("monthly income")
plt.ylabel("expense on food")
plt.title("income vs expense")
plt.plot(salary, food)
plt.show()
```

# Create Scatterplot using seaborn


```python
import csv
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np

with open("data/ex1data1.txt") as csvfile:
    population, profit = zip(*[(float(row['Population']), float(row['Profit'])) for row in csv.DictReader(csvfile)])
    
df = pd.DataFrame()
df['Population'] = population
df['Profit'] = profit

sns.lmplot(x="Population", y="Profit", data = df, fit_reg=False, scatter_kws={'s':45})
```

```python
data_frame = pd.read_csv("data/ex1data1.txt")
sns.lmplot(x="Population", y="Profit", data = data_frame, fit_reg = True, scatter_kws={'s':45})
```
