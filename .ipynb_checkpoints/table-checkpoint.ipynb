```python
def table(n = 10):
    f= open("table.csv","w+")
    f.write("i,j,k\n")
    for i in range(n + 1):
        for j in range(11):
            row = str(i) +"," + str(j) + "," + str(i * j)
            f.write(row  + "\n")
            
    f.close()
```

```python
table(1000)
```

# Analyze table

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data_frame = pd.read_csv('table.csv')
```

```python
data_frame
```

```python
from sklearn.cross_validation import train_test_split

featured_column_name = ['i', 'j']
predict_column_name = ['k']

X = data_frame[featured_column_name].values
y = data_frame[predict_column_name].values

split_test_size = 0.3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = split_test_size, random_state = 42)
```

```python
print("{0:0.2f}% in training set".format((len(X_train)/len(data_frame.index)) * 100))
print("{0:0.2f}% in test set".format((len(X_test)/len(data_frame.index)) * 100))
```

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(X_train, y_train.ravel())
prediction_nb = nb_model.predict(X_test)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, prediction_nb)

print (accuracy)
```
